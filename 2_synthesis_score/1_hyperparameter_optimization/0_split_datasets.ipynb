```python
import pandas
from led3_score.fingerprints.fingerprints import DrugExFingerprint
from led3_score.fingerprints.fingerprint_generator import FingerprintGenerator

drugex_fingerprint = DrugExFingerprint()
fingerprint_generator = FingerprintGenerator(fingerprint = drugex_fingerprint)
```

```python
caspyrus_50k_result_path = "<PATH>/led3_score/data/results/1_route_planning/caspyrus_50k_results.csv"
caspyrus_50k_results = pandas.read_csv(caspyrus_50k_result_path)

# one can find more than one route, so we set everything above than 1 found route to a True value
caspyrus_50k_results["Y_led3"] = caspyrus_50k_results["led3_result"] > 0
caspyrus_50k_results["Y_zinc"] = caspyrus_50k_results["zinc_result"] > 0

caspyrus_50k_results = fingerprint_generator.create_fingerprints_pandas_dataframe(data_frame = caspyrus_50k_results, smiles_column = "clean_smiles", fingerprint_column = "fingerprint")
caspyrus_50k_results = fingerprint_generator.flatten_fingerprints_in_dataframe(dataframe = caspyrus_50k_results, fingerprint_column = "fingerprint")
caspyrus_50k_results.head()
```

```python
chembl_200k_random_sample_path = "<PATH>/led3_score/data/results/1_route_planning/chembl200k_results.csv"
chembl_200k_random_sample_results = pandas.read_csv(chembl_200k_random_sample_path)

# one can find more than one route, so we set everything above than 1 found route to a True value
chembl_200k_random_sample_results["Y_led3"] = chembl_200k_random_sample_results["led3_result"] > 0
chembl_200k_random_sample_results["Y_zinc"] = chembl_200k_random_sample_results["zinc_result"] > 0

chembl_200k_random_sample_results = fingerprint_generator.create_fingerprints_pandas_dataframe(data_frame = chembl_200k_random_sample_results, smiles_column = "clean_smiles", fingerprint_column = "fingerprint")
chembl_200k_random_sample_results = fingerprint_generator.flatten_fingerprints_in_dataframe(dataframe = chembl_200k_random_sample_results, fingerprint_column = "fingerprint")
chembl_200k_random_sample_results.head()
```

```python
def create_splits(dataframe, path, dataset_name, stratify_column):

    # split into train and test with 90/10 ratio sklearn split
    from sklearn.model_selection import train_test_split
    train, test = train_test_split(dataframe, test_size=0.1, random_state=42, stratify=dataframe[stratify_column])

    dataframe.to_csv(f"{path}/{dataset_name}.csv", index=False)
    train.to_csv(f"{path}/{dataset_name}_train.csv", index=False)
    test.to_csv(f"{path}/{dataset_name}_test.csv", index=False)
```
