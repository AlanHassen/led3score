```python
known_ligands = "<PATH>/led3_score/data/chemical_datasets/target_datasets/Q99685_papyrus_known_ligands/Q99685_papyrus_low.tsv"

import pandas as pd

known_ligands_df = pd.read_csv(known_ligands, sep="\t")
known_ligands_df.head()
```

```python
# get the rows with the highest pchembl_value_Median score
known_ligands_df.sort_values(by="pchembl_value_Median", ascending=False).head()
```

```python
# visualize the distribution of the pchembl_value_Median
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(known_ligands_df["pchembl_value_Median"])
plt.show()
```

```python
# canonicalize the smiles

from rdkit import Chem

def canonicalize_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return Chem.MolToSmiles(mol)

known_ligands_df["clean_smiles"] = known_ligands_df["SMILES"].apply(canonicalize_smiles)
known_ligands_df.head()
```

```python
# visualize smiles

from rdkit.Chem import Draw

def visualize_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return Draw.MolToImage(mol)


visualize_smiles(known_ligands_df["SMILES"][699])

```

```python
visualize_smiles(known_ligands_df["clean_smiles"][699])
```

```python
# create inchi keys

from rdkit.Chem import rdinchi

def get_inchi_key(smiles):
    mol = Chem.MolFromSmiles(smiles)
    inchi = Chem.MolToInchi(mol)
    inchi_key = rdinchi.InchiToInchiKey(inchi)
    return inchi_key

known_ligands_df["clean_smiles_inchi_key"] = known_ligands_df["clean_smiles"].apply(get_inchi_key)
known_ligands_df.head()
```

```python
# rows where InChIKey and  clean_smiles_inchi_key differ

known_ligands_df[known_ligands_df["InChIKey"] != known_ligands_df["clean_smiles_inchi_key"]]
```
