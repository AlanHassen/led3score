# Determine Distributions of Scores

Uses the output file from [FT_analysis.ipynb](FT_analysis.ipynb) to determine initial distributions of predictor scores before RL and also saves the scored data to a file to be reused in [RL_analysis.ipynb](RL_analysis.ipynb) for comparison after RL.

```python
df_all = pd.read_table(GENERATED_FT_PATH_ALL, sep='\t', header=0)
df_all
```

```python
from predictors import get_led3, get_qsar, get_qsar_reg, get_sascore, run_predictors

# load predictors/scorers
led3_path = "models/sa_predictors/led3_chembl200k_random_hist_xgboost.model"
led3_scorer = get_led3(led3_path, False)
sascore_scorer = get_sascore(False)
qsar_scorer = get_qsar(False)
qsar_scorer_reg = get_qsar_reg()
```

```python
# sanity check on known molecules from training data
test_mols = [
    "Cc1ccc(C)n1Cc1ccc(-c2nn(C(=O)N(C)C)nn2)cc1", # active
    "CCCCCCCCCC(=NO)c1c(O)ccc(-c2ccc(OC)cc2)c1" # inactive
]
predictors = [led3_scorer, qsar_scorer, qsar_scorer_reg]
run_predictors(test_mols, predictors=predictors)
```

```python
df_all_scores = run_predictors(df_all.SMILES, predictors=predictors)
df_all_scores
```

```python
predictor_keys = [x.getKey() for x in predictors]
predictor_keys
```

```python
# append scores to the file with all the data
df_all = pd.concat([
    df_all.loc[:,~df_all.columns.isin(predictor_keys)],
    df_all_scores[predictor_keys]
], axis=1)
df_all.to_csv(GENERATED_FT_PATH_ALL, sep='\t', header=True, index=False)
df_all
```

```python
# quick overview of simple statistics
for idx, group in df_all.groupby("Set"):
    print(group.Set.iloc[0])
    print(group[predictor_keys].describe())
```
