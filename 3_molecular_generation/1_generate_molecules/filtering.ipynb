```python
from drugex.data.datasets import GraphFragDataSet

train = GraphFragDataSet(f"data/encoded/Q99685_train.tsv", rewrite=False)
test = GraphFragDataSet(f"data/encoded/Q99685_test.tsv", rewrite=False)
```

```python
from drugex.data.corpus.vocabulary import VocGraph

def decode(ds):
    voc = train.getVoc()
    loader = ds.asDataLoader(512)
    frags, smiles = [], []             
    for src in loader:
        f, s = voc.decode(src)
        frags += f
        smiles += s
        
    return frags, smiles

frags, smile = decode(train)
df = pd.DataFrame({"Frag" : frags, "SMILES" : smile})
df.to_csv("data/encoded/Q99685_train_decoded.csv")
df
```

```python
frags_train, smile_train = decode(train)
frags_test, smile_test = decode(test)
```
