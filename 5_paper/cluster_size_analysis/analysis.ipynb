```python
import numpy as np

def count_elements(element):
    if element is np.nan:
        return 0
    else:
        return len(element.split())

count_elements(butina_cluster["Clustermember"][90000])
```

```python
butina_cluster["number_of_cluster_member"] = butina_cluster["Clustermember"].apply(count_elements)
butina_cluster["number_of_cluster_member"] 
```

## count the amount of molecules in each cluster

```python
def get_number(caspyrus_inchi_key):
    temp_row = butina_cluster.loc[butina_cluster['Centroid'] == caspyrus_inchi_key]
    return temp_row["number_of_cluster_member"].values[0]

get_number(full_caspyrus["InChIKey"][6])
```

```python
full_caspyrus["number_of_cluster_member"] = full_caspyrus.apply(lambda row: get_number(row["InChIKey"]), axis=1)
full_caspyrus
```
