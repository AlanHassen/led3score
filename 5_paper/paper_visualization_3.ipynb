## candidate plot


```python
all_results = "<PATH>/led3_score/data/results/3_molecular_generation/generated_molecules/full_model/used_scorer/caspyrus10k_led3/generated_molecules/Q99685_casp10k_led3score_400ep_final_cleaned.csv"

import pandas

casp10k_led3score = pandas.read_csv(all_results)
casp10k_led3score
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming casp10k_led3score is a DataFrame with your data
data = casp10k_led3score[["LED3Scorer", "XGBClassifier"]]

# Set Seaborn style
sns.set(style="white")

# Define the "muted" color palette
palette = sns.color_palette("muted")

# Plotting the data and the Pareto front using Seaborn without legend and without outlines
plt.figure(figsize=(6, 6))
plot = sns.jointplot(x=data["LED3Scorer"], y=data["XGBClassifier"], 
    color=palette[0], s=2, alpha=0.8, legend=False, linewidths=0)
plt.xlabel('In-house synthesizability score')
plt.ylabel('Qsar')
#plt.title('Possible Drug Candidates', fontsize=16)


plt.tight_layout()
plt.savefig("<PATH>/led3_score/5_paper/svgs/experiment_3_led3caspyurs_pareto_candidates.svg", bbox_inches='tight', format="svg")
plt.savefig("<PATH>/led3_score/5_paper/pdfs/experiment_3_led3caspyurs_pareto_candidates.pdf", bbox_inches='tight', format="pdf")

plt.show()
```
