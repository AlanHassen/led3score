```python
from rdkit import Chem
from rdkit import DataStructs
from rdkit.Chem import AllChem

def tanimoto_distance(smiles1, smiles2):
    # Convert the SMILES strings to RDKit molecule objects
    mol1 = Chem.MolFromSmiles(smiles1)
    mol2 = Chem.MolFromSmiles(smiles2)
    
    # Check if the molecules are valid
    if not mol1 or not mol2:
        raise ValueError("Invalid SMILES string")
    
    # Generate Morgan fingerprints for the molecules
    fp1 = AllChem.GetMorganFingerprintAsBitVect(mol1, 2, nBits=2048)
    fp2 = AllChem.GetMorganFingerprintAsBitVect(mol2, 2, nBits=2048)
    
    # Calculate the Tanimoto similarity
    tc = DataStructs.TanimotoSimilarity(fp1, fp2)
    
    return tc

candidate_similarity_0 = tanimoto_distance(selected_candidates["target"][0], selected_candidates["closest_ligand"][0])
print(f"Similarity candidate - target: {candidate_similarity_0}")#

candidate_similarity_1 = tanimoto_distance(selected_candidates["target"][1], selected_candidates["closest_ligand"][1])
print(f"Similarity candidate - target: {candidate_similarity_1}")

candidate_similarity_2 = tanimoto_distance(selected_candidates["target"][2], selected_candidates["closest_ligand"][2])
print(f"Similarity candidate - target: {candidate_similarity_2}")
```
